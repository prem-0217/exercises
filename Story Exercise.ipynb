```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
food = pd.read_csv('indian_food.csv')
food.head()
```

```python
food.shape
```

```python
food.loc[110,['region']] = 'North'
food['region'] = food['region'].str.replace('-1','all')
food.drop(food[food.region=='Central'].index, inplace=True)
```

```python
_ = food['region'].value_counts().plot(kind='pie')
_.set_title('Distribution of recipes in India by region')
plt.show()
```

```python
_ = food['course'].value_counts().plot(kind='pie')
_.set_title('Distribution of recipes in India by course')
plt.show()
```

```python
dessert = food[food.course == 'dessert']
main = food[food.course == 'main course']
```

```python
colors = ['m','b','y','r','g','c']
_ = dessert['region'].value_counts().plot(kind = 'bar', color= colors)
_.set_title('Amount of Desserts by Region')
```

```python
colors = ['r', 'y', 'm', 'g', 'b', 'c']
_ = main['region'].value_counts().plot(kind = 'bar', color= colors)
_.set_title('Amount of Main Courses by Region')
```

```python
colors = ['b','b','r','r','r','r','g','g','y','y','y','m','m','m','c','c','c']
_ = food.groupby('region')['course'].value_counts().plot(kind='bar', color=colors)
_.set_title('Overview of food types by region')
```

```python
_ = food.groupby('region').mean().plot(kind='bar')
_.set_title('Regional Cook and Prep time Averages')
```
